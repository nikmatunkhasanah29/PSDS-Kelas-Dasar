**Looping/Perulangan**

Python menyediakan 2 statement untuk melakukan proses perulangan yaitu for and while. Secara umum perulangan for lebih banyak digunakan daripada while.  

**For Loop**

Perulangan for digunakan untuk melakukan iterasi di dalam sequance seperti list, tuple, string,dan array atau object yang dapat diiterasi lainnya. 

For Loop syntax

```
for val in sequence :
    loop body
```

**For Loop Flowchart**



Capture.PNG

````python
#Contoh for loop
#Program untuk menghitung semua angka dalam sebuah list
L1 = [2, 4, 5, 7, 9]
jumlah = 0

#Iterasi didalam list
for i in L1:
  jumlah = jumlah + i
print(jumlah) 
````

**The range() function**

````python
#program untuk mlakukan iterasi didalam list melalui indeks
minuman = ["teh","boba","kopi","susu"]
p = len(minuman)

#iterasi didalam list melalui indeks
for i in range(p):
  print(minuman[i]) 
````

**For Loop with else**

````python
#Contoh for loop menggunakan else
for i in range (p):
  print(minuman[i])
else:
  print("Elemen sudah habis") 
````

**While Loop**

Digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true. 

Syntax of while loop in python

```
while test_expression:
     Body of while
```

**While Loop Flowchart**

Capture1.PNG

````python
#Contoh
#Program untuk menjumlahkan semua bilangan asli hingga suku ke-n
#Jumlah = 1 + 2 + 3 + ... + n

n = 10
jumlah = 0
i = 1

while i <= n:
  jumlah = jumlah + i
  i = i + 1

print(jumlah)
````

**While Loop With else**

Sama dengan **for loop** kita juga dapat mengkombinasikan **while loop** dengan **blok else**. Bagian **else** akan dieksekusi jika nilai pada **test expression** pada **while loop** bernilai **False**. 

````python
#Contoh
jum = 0
i = 1

while i <= n:
  jum = jum + i
  i = i + 1
else:
  print(jum)  #dieksekusi diluar looping
````

**LATIHAN 1**

Diberikan **list_tagihan** diminta untuk menghitung jumlah seluruh tagihan. Gunakan looping **for** untuk menyelesaikannya. Simpan hasilnya pada variabel **total_tagihan** 

```
list_tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
``` 

````python
#Penyelesaian:
tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
total_tagihan = 0

#Iterasi didalam list
for i in tagihan:
  total_tagihan = total_tagihan + i

print(total_tagihan) 
````

**LATIHAN 2**

Diberikan list **tagihan** hitunglah **total_tagihan** menggunakan **while loop** 

```
tagihan = [50000, 75000, 125000, 300000, 200000]
```

````python
#Penyelesaian:
tagihan = [50000, 75000, 125000, 300000, 200000]
jumlah = len(tagihan)
total_tagihan = 0
i = 0

#Menggunakan perulangan while
while i < jumlah:
  total_tagihan += tagihan [i]
  i = i + 1

print(total_tagihan)
````
