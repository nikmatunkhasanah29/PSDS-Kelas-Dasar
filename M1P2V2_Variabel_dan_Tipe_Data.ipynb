Variabel dan Type Data

Variabel adalah lokasi penyimpanan yang dapat digunakan untuk menampung sebuah data atau informasi. Syarat-syarat penamaan variabel antara lain:
1. Karakter-karakter yang digunakan untuk penamaan variabel adalah alphabet, angka dan underscore.
2. Karakter pertama variabel harus berupa huruf atau garis bawah/underscore dan tidak bisa berupa angka.
3. Nama sebuah variabel tidak bisa menggunakan keyword atau reserved words dari bahasa python seperti for, if dan lainnya.
4. Karakter pada nama variabel bersifat sensitif (case-sensitif)

````python
#Membuat variabel
kalimat = 'ini adalah variabel'
bilangan1 = 4
bilangan2 = 2.5
bilangan3 = False
bilangan4 = 1j+4
````

````python
#Menampilkan variabel
print(kalimat)
print(bilangan1)
print(bilangan2)
````

Tipe Data

Lima tipe data yaitu:
1. Integer atau bilangan bulat
2. Float atau bilangan desimal
3. String atau kalimat
4. Boolean yang terdiri dari True dan False
5. Compleks yang merupakan bilangan kompleks

````python
#Menampilkan variabel
print(type(kalimat))
print(type(bilangan1))
print(type(bilangan2))
print(type(bilangan3))
print(type(bilangan4))
````

Konversi Type Data

Syntax konversi type data yaitu:
1. Integer = int()
2. Float   = float()
3. String  = str()
4. Boolean = bool()
5. Complex = complex()

````python
#Konversi type data integer
print(int(bilangan2))
print(int(bilangan3))
````

````python
#Konversi type data float
print(float(bilangan1))
print(float(bilangan3))
````

````python
#Konversi type data string
print(str(bilangan1))
print(str(bilangan2))
print(str(bilangan3))
````

````python
#Konversi type data bool
print(bool(bilangan1))
print(bool(bilangan2))
````

````python
#Konversi type data complex
print(complex(bilangan1))
print(complex(bilangan2))
print(complex(bilangan3))
````
