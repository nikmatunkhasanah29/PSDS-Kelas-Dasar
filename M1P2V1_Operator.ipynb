Operator

Operator adalah simbol yang dapat digunakan untuk mengubah nilai dari satu variabel dengan melibatkan satu atau lebih variabeldan literal.
Tipe-tipe operator:
1. Operator aritmatika
2. Operator assignment
3. Operator pembanding
4. Operator logika
5. Operator membersip

Operator Aritmatika digunakan untuk melakukan operasi matematika yang umum.
Macam-macam operator aritmatika antara lain:

```
+ : Penjumlahan
- : Pengurangan
* : Perkalian
/ : Pembagian
% : Modulus
**: Perpangkatan
//: Pembagian dengan hasil pembulatan flooring
```

````python
#Contoh penggunaan operator aritmatika
#Penjumlahan
print('Penjumlahan')
print(2+2)

#Pengurangan
print('-'*15)
print('Pengurangan')
print(4-3)

#Perkalian
print('-'*15)
print('Perkalian')
print(5*5)

#Pembagian
print('-'*15)
print('Pembagian')
print(70/7)

#Modulus
print('-'*15)
print('Modulus')
print(2%3)

#Perpangkatan
print('-'*15)
print('Perpangkatan')
print(2**3)

#Pembagian pembulatan floring
print('-'*15)
print('Pembagian pembulatan floring')
print(3//2)
````

Operator Assigment digunakan untuk menyimpan nilai menjadi suatu variabel.

```
=   : contohnya x = 5       -> x  = 5
+=  : contohnya x = x + 5   -> x += 5
-=  : contohnya x = x - 5   -> x -= 5
*=  : contohnya x = x * 5   -> x *= 5
/=  : contohnya x = x / 5   -> x /= 5
%=  : contohnya x = x % 5   -> x %= 5
//= : contohnya x = x // 5  -> x //= 5
**= : contohnya x = x ** 5  -> x **= 5
```

Operator Pembanding digunakan untuk membandingkan 2 nilai atau lebih variabel.

Macam-macam operator pembanding antara lain:

```
==  -> sama dengan
!=  -> tidak sama dengan 
>   -> lebih dari
<   -> kurang dari
>=  -> lebih dari dan sama dengan
<=  -> kurang dari dan sama dengan
```

````python
#Contoh penggunaan operator pembanding
print(2==2)
print(2!=2)
print(2>2)
print(2<2)
print(2>=2)
print(2<=2)
````

Operator Logika digunakan untuk mengkombinasikan statement-statement kondisional. 

Macam-macam operator logika antara lain:

```
and -> kondisi menjadi benar apabila semua statementnya benar
or  -> kondisi menjadi benar apabila salah satu statementnya ada yang benar
not -> membalik hasildari benar ke salah dan salah ke benar
```

````python
#Contoh penggunaan operator logika
print(True and False)
print(False or False)
print(not False and True)
````

Operator Keanggotaan digunakan untuk menguji apakah urutan disajikan dalam satu objek 

Macam-macam operator keanggotaan antara lain:

```
in      -> return True ketika nilai tertentu dalam sebuah urutan ada
not in  -> return True ketika nilai tertentu dalam sebuah urutan tidak ada
```

````python
#Contoh penggunaan operator keanggotaan
a = (1,2,3,4,5,6)
b = 2
print(b in a)

print('-'*15)
a = (1,2,3,4,5,6)
b = 7
print(b not in a)
````
